```python
windowSize = 11
def labl(df,windowSize =11):
    labels = []
    values = []
    for i in range(len(df.close) - windowSize):
        mx = df.close[i]
        mn = df.close[i]
        mxIndex, mnIndex = i, i

        for j in range(i + 1, i + windowSize + 1):
            if df.close[j] > mx:
                mx = df.close[j]
                mxIndex = j
            if df.close[j] < mn:
                mn = df.close[j]
                mnIndex = j

        if mnIndex == i + 6:
            labels.append(1)
            values.append(i + 6)
        elif mxIndex == i + 6:
            labels.append(0)
            values.append(i + 6)
        else:
            labels.append(2)
            values.append
    return labels,values
```

```python
import keras_tuner as kt
from keras_tuner.tuners import RandomSearch
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Dropout, Flatten
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import categorical_crossentropy
```

```python
class MyHyperModel(kt.HyperModel):
    def build(self, hp):
        model = Sequential()
        model.add(Conv2D(
            filters=hp.Int('conv1_filters', min_value=4, max_value=64, step=4),
            kernel_size=hp.Int('conv1_kernel', min_value=4, max_value=64, step=2),
            padding='same',
            input_shape=(1, 9, 9),
            activation='relu'
        ))
        model.add(MaxPooling2D(pool_size=(2, 2), padding='same'))
        model.add(Dense(
            units=hp.Int('dense1_units', min_value=64, max_value=512, step=4),
            activation=hp.Choice('dense1_activation', values=['relu', 'sigmoid', 'tanh'])
        ))
        model.add(Dropout(rate=hp.Float('dropout1', min_value=0.1, max_value=0.5, step=0.01)))
        model.add(Flatten())
        model.add(Dense(
            units=hp.Int('dense2_units', min_value=64, max_value=512, step=50),
            activation=hp.Choice('dense2_activation', values=['relu', 'sigmoid', 'tanh'])
        ))
        model.add(Dense(3, activation='softmax'))

        model.compile(
            loss='categorical_crossentropy',
            optimizer=Adam(learning_rate=hp.Choice('learning_rate', values=[1e-2, 1e-3, 1e-4])),
            metrics=['accuracy']
        )

        return model
```

```python
tuner = kt.RandomSearch(
    MyHyperModel(),
    objective='val_accuracy',
    max_trials=10,
    directory='my_tuner_directory',
    project_name='my_cnn_tuner'
)
```

```python
tuner.search(X_a, y_train, epochs=10, validation_data=(X_b, y_t))
```

```python
best_hyperparameters = tuner.get_best_hyperparameters(num_trials=1)[0]
best_model = MyHyperModel().build(best_hyperparameters)
```

```python
best_hyperparameters = tuner.get_best_hyperparameters(num_trials=1)[0]
print("Best Hyperparameters:")
print(best_hyperparameters.values)
```

```python
best_model.fit(X_a, y_train, epochs=100, validation_data=(X_b, y_t))
```
